```python
def plot_groups(subgroups, test_id, grade):
    df = p.select("""
SELECT * from star 
WHERE subgroup IN ({subgroups}) 
AND test_id = 7 -- CST English-Language Arts
AND school != 0
AND grade = grade
""".format(test_id=test_id,grade=grade,subgroups=','.join([str(x) for x in subgroups]))).pandas
    sdf = df[["subgroup","mean_score"]].dropna()
    
    colors = 'k','r','b','g','m','c'
    
    for i, sg in enumerate(subgroups):
        sdf[sdf['subgroup'] == sg]['mean_score'].plot(kind='kde', color=colors[i])
        #sdf[sdf['subgroup'] == sg]['mean_score'].hist(bins=100,color=colors[i])
    
    
```

```python
plot_groups([200,220,206,226], 7, 8)
```

```python
plot_groups([200,220,206,226], test_id=23, grade=10)
```

```python
# Parent Education
plot_groups([90,91,92,93,94], 7, 8)
```

```python
# AA, Hisp, Fil, white, Asian
plot_groups([220,224,223,226,222], 7, 8)
```

```python
# All students vs gifted
plot_groups([1, 30], 7, 8)
```

```python
# Male Vs Female
plot_groups([3,4], 7, 8)
```

```python
# Male Vs Female
plot_groups([3,4], 8, 8)
```

```python
# Economic Status
plot_groups([31,111], 7, 8)
```
